```python
import pandas as pd
import os
```

```python
file_list = os.listdir('data/Rainier Oregon Crashes in XSD and XML')
xml_files = [file for file in file_list if file.endswith('.xml')]
```

```python
xml_files
```

```python
pd.read_xml('data/Rainier Oregon Crashes in XSD and XML/CRASH.xml')
```

```python
files_loc = 'data/Rainier Oregon Crashes in XSD and XML'

for i in range(len(xml_files)):
    current_file = xml_files[i]
    print('{} turned into csv'.format(current_file))
    pd.read_xml('{}/{}'.format(files_loc, current_file)).to_csv('{}/{}.csv'.format(files_loc, current_file[:-4]), index=False)
```
